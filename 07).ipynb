```python
import pandas as pd
import random

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

data = []

for _ in range(300):
    data.append([
        random.randint(6, 72),
        random.choice(['good', 'bad', 'perfect', 'critical']),
        random.randint(1, 4),
        random.randint(500, 20000),
        random.choice(['unemployed', '1-4 years', '4-7 years', '7+ years']),
        random.choice(['none', 'guarantor', 'co-applicant']),
        random.choice(['good', 'bad'])
    ])

dataset = pd.DataFrame(data, columns=[
    'duration',
    'credit_history',
    'existing_credits',
    'credit_amount',
    'employment',
    'other_parties',
    'class'
])

label_encoder = LabelEncoder()

for column in dataset.columns:
    if dataset[column].dtype == 'object':
        dataset[column] = label_encoder.fit_transform(
            dataset[column].astype(str)
        )

X = dataset.drop('class', axis=1)
y = dataset['class']

decision_tree = DecisionTreeClassifier()

cv_scores = cross_val_score(
    decision_tree,
    X,
    y,
    cv=5
)

decision_tree.fit(X, y)

print(decision_tree)

print("\nCross-Validation Accuracy Scores:")
print(cv_scores)

print("\nAverage Accuracy:")
print(cv_scores.mean())
```
